1. What is a random variable in probability theory?
 - A random variable is a numerical outcome of a random experiment. It maps the outcomes of a random process to real numbers.

2. What are the types of random variables?
 - There are two main types:
 - Discrete Random Variable: Takes countable values (e.g., number of students).
 - Continuous Random Variable: Takes any value within a range (e.g., height, weight).

3. What is the difference between discrete and continuous distributions?
 - Discrete Distribution: Deals with countable outcomes (e.g., binomial, Poisson).
 - Continuous Distribution: Deals with infinite possible values within a range (e.g., normal, exponential).

4. What are probability distribution functions (PDF)?
 - A PDF gives the probability of a random variable taking on a particular value. For continuous variables, it represents the density rather than exact probability.

5. How do cumulative distribution functions (CDF) differ from PDF?
 - PDF shows the probability density at a specific value.
 - CDF shows the probability that a variable is less than or equal to a certain value.

6. What is a discrete uniform distribution?
 - A discrete uniform distribution assigns equal probability to a finite number of outcomes.
Example: Rolling a fair die (each face has 1/6 probability).

17. What are point estimates and interval estimates in statistics?
 - Point Estimate: A single value estimate of a parameter (e.g., sample mean).
 - Interval Estimate: A range (confidence interval) that likely contains the population parameter.

18. What is the significance of confidence intervals in statistical analysis?
 - They provide a range within which we expect the population parameter to lie with a certain level of confidence (e.g., 95%).
